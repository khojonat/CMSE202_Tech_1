**Image Zoning**

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import collections
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from IPython.display import Image
from sklearn.datasets import fetch_openml
```

```python
def cos_similarity(v1,v2):
    '''
    returns cosine similarity between two vectors
    '''
    return np.dot(v1,v2)/np.dot(np.linalg.norm(v1),np.linalg.norm(v2))

Image(filename='Cosine-Similarity-Formula-1.png')
```

```python
digits = load_digits()
mnist = fetch_openml('mnist_784')
mnist_array = np.array(mnist.data)
mnist_array = mnist_array.reshape((70000, 28, 28))
#Loading in and reshaping mnist data set
```

```python
plt.imshow(mnist_array[69998,:,:], cmap='gray')
#Displaying 5
```

```python
def light_zone_Vector(array):
    '''
    -vec is a list that holds the zone vector
    -cnt for the selected rows counts the amount of light pixels from the 1-14th columns
    -cnt1 cnt for the selected rows counts the amount of light pixels from the 15-28th columns
    -extra adds amount of rows after each iteration to change zones to count new pixels
    
    The light_zone_Vector function counts the amount of light pixels for each zone in a mnist image. Returns the vector as a numpy array.
    '''
    vec=[]
    cnt=0
    cnt1=0
    extra=0
    for i in range(0,7):
        cnt=0
        cnt1=0
        for row in range(0+extra,4+extra):
            for col in range(0,14):
                if array[row][col]>110:
                    cnt+=1
            for col in range(14,28):
                if array[row][col]>110:
                    cnt1+=1
        vec.append(cnt)
        vec.append(cnt1)
        extra+=4

    return np.array(vec)
```

```python


v1=light_zone_Vector(mnist_array[0,:,:])

v2=light_zone_Vector(mnist_array[69998,:,:])

print(v1,"\n",v2)
print(cos_similarity(v1,v2))
```

```python
train_vectors1, test_vectors1,train_labels1, test_labels1 = train_test_split(mnist_array,list(int(i) for i in mnist.target), test_size=0.8, random_state=2)
```

```python
def zone_vectors_for_classification(traning_set,training_class_labels):
    '''
    Takes training set and training labels as parameters
    
    -light is a dictionary that holds the average zone vector of each class in the training set
    -cnt is a count of how many digits there are in the training set for each class
    
    The zone_vectors_for_classification is a function that returns a dictionary of zone vectors for each class of the training set.
    The zone vectors are the average of every zone vector within that particular class.
    '''
    light={}
    cnt=collections.Counter(training_class_labels)
    
    for i in range(len(traning_set)):
        if training_class_labels[i] not in light:
            light[training_class_labels[i]]=light_zone_Vector(traning_set[i])
        else:
            light[training_class_labels[i]]+=light_zone_Vector(traning_set[i])
            
    for key in light:
        light[key]=light[key]/cnt[key]
    return light
def predict(lightzone_vec_classifer,image):
    '''
    Takes a light zone classifier and image as input.
    - max1 is the maximum cosine similarity value
    - classification is the class's zone vector that is most similar to image input
    The predict function is used to predict the class of the image using the lightzone_vec_classifier.
    '''
    max1=-1
    classification=-10000
    light_image_vec=light_zone_Vector(image)
    for key in lightzone_vec_classifer:
        if cos_similarity(light_image_vec,lightzone_vec_classifer[key])>max1:
            max1=cos_similarity(light_image_vec,lightzone_vec_classifer[key])
            classification=key
    return classification
    
```

```python
light_zone_vecs_classifiers1=zone_vectors_for_classification(train_vectors1,train_labels1)
#Training classifier
```

```python
predict_vectors1 = test_vectors1
true_labels1 = test_labels1

pred_labels1 = [predict(light_zone_vecs_classifiers1,image) for image in test_vectors1]

print(classification_report(true_labels1, pred_labels1))
print(confusion_matrix(true_labels1, pred_labels1, labels=range(10)))
```
